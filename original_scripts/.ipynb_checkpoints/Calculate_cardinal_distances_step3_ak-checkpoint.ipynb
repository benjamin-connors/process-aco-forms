```python
#set paths and import files and output file to export
path = r"G:\ACO\2024"
#file = path+ "\\1_scripts\working\p1\TSI_23_P01.csv"
file = path + "\\1_scripts\processing_data\DM_trip" + str(flt_no) + "_clean_cru.csv"
template = "templates\cardinal_dir_template.csv"
#gnss_temp = "templates\Point_template.xlsx"
# eng_gnss = path + r"\Englishman\3_GNSS\2_Working\P1\GNSS_EGM_P1_unfixed_wo.csv"
cru_gnss = path + r"\Cruickshank\3_GNSS\2_Working\P1\GNSS_cru_phase1_points.csv"
#rus_gnss = path + r"\Tsitika\3_GNSS\2_Working\GNSS_TSI_P1_unfixed.csv"
# mv_gnss = path + r"\Metro_Van\3_GNSS\2_Working\P1\GNSS_MV.csv"
#export to
#output = "processing_data\calc_cardinal_trip1"

#read the field data, and template for adding distances
#ensure field types are correct for later merging
df = pd.read_csv(file, encoding = 'mac-roman', dtype={"cardinal": "object", "distance_m": "float64", "plot_id":"object"})
template = pd.read_csv(template, dtype={"cardinal": "object", "distance_m": "float64"})
#rus_gnss = pd.read_csv(rus_gnss, dtype={"plot_id": "object"})
df = df[df["cardinal"].notna()]
#display(df)
print("# of row:"  + str(df.shape[0]))
#display(template)
```

```python
df.plot_id.unique()
```

```python
#drop rows that have Nan values in Study Area
df.dropna(subset=["study_area"])
```

```python
#filter by cardinal plots assign to cardinal variable and to road transect variable. 
cardinal = df[df["plot_type"]=="Cardinal 10 m"]
road = df[df["plot_type"]== "Road Transect"]

#set apart the remaining dataframe
other = df[(df["plot_type"] !="Cardinal 10 m") | (df["plot_id"] !="Road Transect")]

#print out the number of rows for both cardinal and road variables. Check that against total to check for reference after merge
print("# of rows in cardinal df:" + str(cardinal.shape[0]))
print("# of rows in road df:" + str(road.shape[0]))
```

```python
#set up for merge
key = ["cardinal", "distance_m"]
cardinal = cardinal.merge(template, on=key, suffixes=("_1", "_2"))
road = road.merge(template, on=key, suffixes=("_1", "_2"))
```

```python
#gnss= pd.read_csv(str(study_area) + "_gnss")
#gnss= pd.read_csv(eng_gnss, encoding = 'mac_roman')
#gnss= pd.read_csv(rus_gnss)
gnss= pd.read_csv(cru_gnss)
#rus_gnss = pd.read_csv(rus_gnss, dtype={"plot_id": "object"})
# gnss= pd.read_csv(mv_gnss, encoding = 'mac_roman')
```

```python
#gnss['plot_id']= gnss['plot_id'].astype(str)
#df['plot_id']=df['plot_id'].astype(str)
gnss['Northing_m'] = gnss['Northing_m'].astype(float)
gnss['Easting_m'] = gnss['Easting_m'].astype(float)
```

```python
#a) groupby study area
#b) merge with the appropriate GNSS data form
#cru_df =df[df['study_area'] == 'Cruickshank']
# eng_df = df[df['study_area'] == 'Englishman']
#rus_df = df[df['study_area'] == 'Russell Creek']

#cru_df=pd.merge(cru_df, gnss, on="plot_id")
#eng_df=pd.merge(eng_df, gnss, on="plot_id")
#rus_df=pd.merge(rus_df, gnss, on="plot_id")
cru_df=pd.merge(df, gnss, on="plot_id", how = "left")
#cru_df
df.to_csv(path + r"\1_scripts\working\cru_gnss_merged_phase_" + str(flt_no) + ".csv", index=False)
#eng_df.to_csv(path + r"\1_scripts\working\eng_gnss_merged_phase_" + str(flt_no) + ".csv", index=False)
#df.to_csv(path + r"\1_scripts\working\mv_gnss_merged_phase_" + str(flt_no) + ".csv", index=False)
#rus_df.to_csv(path + r"\1_scripts\working\rus_gnss_merged_phase_" + str(flt_no) + ".csv", index=False)
#rus_df
#print("# of row:"  + str(rus_df.shape[0]))
```
